```python
#process_1 ％の削除
def StripPercentSign(text: pd.Series) -> pd.Series:
    return text.replace('%', '',)

#process_2 従業員数から「人」を抜く処理
def StripHuman(text: pd.Series) -> pd.Series:
    return text.replace('人', '') 

#process_3 資本金から「千円」を抜く処理
def StriThousandYen(text: pd.Series) -> pd.Series:
    return text.replace('千円', '')

#process_5 全角処理
def ChangeZenkaku(text: pd.Series) -> pd.Series:
    return text.apply(mojimoji.han_to_zen)

#process_6 不要項目の削除
def DeleteEmpty(text: str) -> str:
    return text.replace(['Unnamed: 0','Unnamed: 1','Unnamed: 2','named: 0'],'')



#（株）,(有)の処理（置き換えstr.replace）
df.iloc[:, 5] = df.iloc[:, 5].str.replace('(株)', '株式会社',regex=False)
df.iloc[:, 5] = df.iloc[:, 5].str.replace('(有)', '有限会社',regex=False)
df.iloc[:, 5] = df.iloc[:, 5].str.replace('(資)','有限会社',regex=False)
df.iloc[:, 5] = df.iloc[:, 5].apply(mojimoji.han_to_zen)
df.iloc[:, 5] = df.iloc[:, 5].str.replace('－','-')



#不要項目を削除
# df= df.replace(['Unnamed: 0','Unnamed: 1','Unnamed: 2','named: 0'],'')

#業種1～3に分類
df['業種1No'] = df['業種1'].str[:5].astype(str).str.zfill(4)
df['業種1'] = df['業種1'].str[5:]
df['業種2No'] = df['業種2'].str[:5]
df['業種2'] = df['業種2'].str[5:]
df['業種3No'] = df['業種3'].str[:5]
df['業種3'] = df['業種3'].str[5:]
```
